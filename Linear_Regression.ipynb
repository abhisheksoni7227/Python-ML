```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore') # hide warning
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.rename(columns={'per capita income (US$)': 'capital income'}, inplace=True)
```

```python
df
```

```python
df.rename(columns={'capital income': 'capital'}, inplace=True)
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('capital')
plt.scatter(df.year,df.capital,color='red',marker='+')
```

```python
new_df=df.drop('capital',axis='columns')
new_df
```

```python
capital=df.capital
capital
```

```python
# Create linear regression object
reg=linear_model.LinearRegression()
reg.fit(new_df,capital)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
